# Data Prep

```python
# menghitung tanggal kapan terakhir transaksi
data['latets_transaction_date'] = data.groupby('Customer Name')['Order Date'].transform('max')
data.head()
```

```python
# melihat pola data time series dengan binders\
data_binders = data[data['Sub-Category'] == 'Binders']

# mengurutkan orderdate
data_binders['Order Date'].min(), data_binders['Order Date'].max()

```

```python
data_binder = data_binders.groupby('Order Date')['Quantity'].sum().reset_index()
data_binder
```

```python
data_binders = data_binders.set_index('Order Date')
data_binders.index
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
data_binders.head()
```

```python
# proses mengurutkan data
data_binders = data_binders['Quantity'].resample('MS').mean()
data_binders
```

```python
data_binders = data_binders.to_frame().reset_index()
data_binders = data_binders.rename(columns = {0 : 'Quantity'})
data_binders.index_name = 'Order Date'
data_binders.set_index('Order Date', inplace = True)
```

```python
data_binders
```

```python
data_binders.sort_index(inplace = True)
data_binders.index
```

```python
dec_data_binders = seasonal_decompose(data_binders['Quantity'], model = 'multiplicative')
dec_data_binders.plot()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
# split data training dan testing
train = data_binders.iloc[:len(data_binders) - 12]
test = data_binders.iloc[:len(data_binders) - 24:]
```

```python
# dalam kasus ini, model SARIMA menggunakan autoregresive (p)
# orde diferencing (d) dan orde moving average (q) masing masing 1,0,0

# seasonal order = 2,1,0,12 : ini adalah parameter seasonal order yang mengacu  pada ordeer komponen ARIMA musiman.
# dalam kasus ini, model SARIMA menggunakan autoregresive (p)
# orde diferencing (d) dan orde moving average (q) dan panjang musiman masing masing 2,1,0,12

model_sarimax = SARIMAX(data_binders['Quantity'], order = (1, 0, 0), seasonal_order = (2, 1, 0, 12))

# training SARIMAX MODEL
result_sarimax = model_sarimax.fit()

# Langkah yang akan di prediksi
forecast_steps = 12

# Prediksi
prediction = result_sarimax.get_forecast(steps=forecast_steps)

# mendapatkan nilai prediksi 
forecast_values = prediction.predicted_mean
confidence_interval = prediction.conf_int()

# menampilkan hasil prediksi
print(forecast_values)
print(confidence_interval)

```

```python
result_sarimax.summary()
```

```python
# menghitung panjang data train dan test
x = len(train)
y = len(train) + len(test) - 1

# prediksi untuk 1 tahun
predict = result_sarimax.predict(x,y, type = 'level').rename('Prediction')

# plot prediksi vs Actual

predict.plot(legend = True)
test['Quantity'].plot(legend = True)
```

```python
#  tools evaluasi
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
from statsmodels.tools.eval_measures import rmse

```

```python
# evaluasi 
mean_absolute_percentage_error(test['Quantity'],predict)
```
