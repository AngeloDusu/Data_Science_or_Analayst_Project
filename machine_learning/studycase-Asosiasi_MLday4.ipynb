# Data Preparation

```python
data['PROVINSI'].value_counts()
```

```python
data_jateng4 = data[data['PROVINSI'] == 'JAWA TENGAH']
data_jateng4
```

```python
data_jateng1 = data[data['PROVINSI'] == 'JAWA TENGAH']
data_jateng2 = data[data['PROVINSI'] == 'Jawa Tengah']
data_jateng3 = data[data['PROVINSI'] == 'JAWATENGAH']
data_jateng4 = data[data['PROVINSI'] == 'JAWA TENGAH ']


data_jateng = pd.concat([data_jateng1, data_jateng2, data_jateng3, data_jateng4], sort = True, axis = 0)
data_jateng.info()
```

```python
data_jateng['PROVINSI'].value_counts()
```

```python
data_jateng['PROVINSI'] = data_jateng['PROVINSI'].replace('JAWA TENGAH', 'Jawa Tengah')
data_jateng['PROVINSI'] = data_jateng['PROVINSI'].replace('JAWATENGAH', 'Jawa Tengah')
data_jateng['PROVINSI'] = data_jateng['PROVINSI'].replace('JAWA TENGAH ', 'Jawa Tengah')
data_jateng['PROVINSI'] = data_jateng['PROVINSI'].replace('Jawa Tengah', 'Jawa Tengah')
```

```python
data_jateng['PROVINSI'].value_counts()
```

1. Menyesuaikan bentuk data supaya bisa di proses oleha algoritma asosiasi -> Apriori
2. Memilih provinsi mana yang akan dimodelkan yaitu Jawa tengah dan sekaligus merapihkan penamaan yang tidak konsisten

```python
# mengubah invoice date kedalam bentuk datetime

data_jateng['InvoiceDate'] = pd.to_datetime(data_jateng['InvoiceDate'])

data_jateng.head(10)
```

```python
# memulai proses transformasi data dari bentuk transaksional ke dalam flat tabel
# tahap 1 :
# mengubah hal yang menjadi prioritas pada keranjang belanja

data_jateng['PRODUCT'] = data_jateng['PRODUCT'].str.strip()
data_jateng['PRODUCT_CATEGORY'] = data_jateng['PRODUCT_CATEGORY'].str.strip()
```

```python
data_jateng.dropna(axis = 0, subset=['InvoiceNo'], inplace = True)
data_jateng['InvoiceNo'] = data_jateng['InvoiceNo'].astype('str')
data_jateng = data_jateng[~(data_jateng['InvoiceNo'].str[0] == ('C'))]
```

```python
data_jateng
```

```python
# Keranjan belanjah

basket_jateng = (data_jateng[data_jateng['PROVINSI'] == 'Jawa Tengah'].groupby(['InvoiceNo', 'PRODUCT_CATEGORY'])['Quantity'].count()\
    .unstack().reset_index().fillna(0).set_index('InvoiceNo'))

basket_jateng
```

```python
# melakukan proses Encoded 

def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
    
basket_jateng_set = basket_jateng.applymap(encode_units)

basket_jateng_set
    
```

# Pemodelan Asosiasi

```python
from apyori import apriori
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import datetime
```

```python
frequent_itemset = apriori(basket_jateng_set, min_support = 0.1, use_colnames = True)
frequent_itemset
```

```python
rule_jateng = association_rules(frequent_itemset, metric = "lift") 
rule_jateng
```

```python
# rule_jateng_tocsv = rule_jateng.to_csv('D:\.vscode\kodingan\DataSource\\rule_jateng.csv')
```

```python
# visualisasi data rule_jateng dengan plotly
rule_jateng['antecedents'] = rule_jateng['antecedents'].apply(lambda x: ', '.join(list(x)))
rule_jateng['consequents'] = rule_jateng['consequents'].apply(lambda x: ', '.join(list(x)))

fig = px.scatter(rule_jateng, x='antecedents', y='antecedent support')
# fig = px.scatter(rule_jateng, x='consequents', y='consequent support')
fig.show()


fig = px.scatter(rule_jateng, x = 'support', y = 'confidence', color = 'lift')
fig

```
